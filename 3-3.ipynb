# 3-3 특성공학과 규제 (p.150)
> 선형 회귀는 특성이 많을 수록 더 정확한 모델
> - 여러 특성을 사용한 다중 회귀에 대해 배우고 사이킷런의 여러 도구를  사용해 봅니다. 복잡한 모델의 과대적합을 막기 위한 릿지 라쏘회귀도 배움. 

### 다중 회귀 
- 서로 다른 특성 여러가지를 같이 사용하는 것
- e.g. 생선의 길이, 무게, 두께 특성을 데이터로 => 선형 회귀가 평면을 학습한다. 

- 기존의 특성을 사용해 새로운 특성을 뽑아 내는 것을 `특성공학` 이라고 한다.

---

```python
import pandas as pd
df = pd.read_csv('data/Fish.csv')
perch_df = df.loc[df['Species'] == 'Perch']
perch_df.head()
```

```python
# 우리의 목표는 무게를 활용해 다른 특성을 알아내는 것 
# 사용할 데이터는 lenght2, height, width

perch_full = perch_df[['Length2', 'Height', 'Width']]
perch_weight = perch_df[['Weight']]
```

```python
# 학습용 데이터와 채점용 데이터로 분류하기
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = \
train_test_split(perch_full, perch_weight)
```

```python
# 이 데이터를 바탕으로 sklearn이 가지고 있는 함수 사용하기
# 사이킷런의 변환기 (함수)

from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(include_bias=False)
#option 값으로 괄호 안에 (include_bias=False) 를 넣으면 이 값을 출력시 array에서 1이 제외된다. 
poly.fit([[3, 5]])
poly.transform([[3, 5]])

# PolynomialFeatures 함수는 
# => 내가 특성을 주면 그 특성으로 만들 수 있는 조합 (경우의 수)을 만들어냄
```

```python
poly = PolynomialFeatures(include_bias = False)
poly.fit(train_input)
# 우리가 가지고 있는 농어의 길이, 높이, 두께를 특성공학으로 바꾸기

train_poly = poly.transform(train_input)
train_poly[:5]
# 보기 좋게 앞에 5개만 잘라서 봐보자
```

```python
train_poly.shape
# 특성공학을 통해 나온 새로운 특성들의 조합 (경우의 수)
# (도미데이터의 개수, 특성공학으로 만들어진 조합의 개수)
```

```python
poly.get_feature_names_out()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(train_poly, train_target)
lr.score(train_poly, train_target)
```

```python
test_poly = poly.transform(test_input)
```

```python
lr.score(test_poly, test_target)
```

`특성공학` 을 통해
기여도가 높은 특성을 => 모델의 성능을 높임. 

a * 길이 + b * 높이 + c * 너비 => 
=> 무게는 높이와 더 관련성이 높을 것 같아~ 
=> a * 길이 + b * 높이 + c * 너비 d * 길이^2 + e * 높이 * 길이 ... etc.

---

```python
# 더 많은 특성을 만들면, 모델의 성능이 더 높아지지 않을까?

poly = PolynomialFeatures(degree=5, include_bias=False)

# 몇 개의 조합으로 만들지 => degree=n개

poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)
train_poly.shape

# 55개의 특성을 만들어냄. 내가 가진 생선 데이터 수보다도 많음.
```

```python
lr.fit(train_poly, train_target)
```

```python
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))

# 컬럼 (특성)이 많아진다고 좋은 건 아니다.
```

## 규제 (p.158)
> 정답 확률을 높여보자. <- `정규화`를 통해
- 훈련 세트를 너무 과도하게 학습하지 못하도록 훼방하는 것
- 모델이 훈련 세트에 과대적합되지 않도록 만드는 것

- 선형 회귀 모델의 경우 특성에 곱해지는 계수 (또는 기울기)의 크기를 작게 만드는 일

- 데이터 자체에 선을 그으면, 경향성을 보여 주지 못함. 훈련 세트의 점수를 낮추고 대신 테스트 세트의 점수를 높여보자.

```python
# 사이킷런에 있는 정규화 
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

ss.fit(train_poly)

train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
```

```python
train_scaled[:2]
# 정규화 과정을 통해, test_scaled 의 array 내 데이터가 변환됨. => 아까는 데이터들을 5제곱 하고 했었음....
```

## 두가지 선형 회귀 - 릿지 ridge 라쏘 lasso

- 선형 회귀 모델에 규제를 추가한 모델
- 릿지는 계수를 제곱한 값을 기준으로 규제를 적용
- 라쏘는 계수의 절댓값을 기준으로 규제를 적용

#### 릿지 회귀
> `from sklearn.linear_model import Ridge`

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
# (괄호에 아무것도 쓰지않은 경우가 기본값;alpha=1.0)
#
#이걸 ridge = Ridge(alpha=0.1) 이렇게 바꾸면 모델의 정확도가 또 달라짐
ridge.fit(train_scaled, train_target)
```

```python
print(ridge.score(train_scaled, train_target))
print(ridge.score(test_scaled, test_target))
# 필요 없다고 생각하는 데이터들을 빼는 것이 `규제`이다. 
```

#### 라쏘 회귀
> `from sklearn.linear_model import Lasso`

```python
from sklearn.linear_model import Lasso

lasso = Lasso()
lasso.fit(train_scaled, train_target)

print(lasso.score(train_scaled, train_target))
print(lasso.score(test_scaled, test_target))
```

```python
# 선을 그리는데, 특성이 너무 많으면 좋지 않으므로, 규제를 하는 것이 => 릿지, 라쏘 회귀이다. 
```

```python
lasso.coef_
# 무수히 많은 0이 발견되는 걸 볼 수 있음
# 라쏘 모델을 유용한 특성만을 남기는 것으로 쓴다. => 불필요한 것을 0으로 만듦. 
```
