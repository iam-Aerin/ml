âœ” **ë¡œê·¸ í•¨ìˆ˜ì˜ íŠ¹ì„±ìƒ ìŒìˆ˜ë¥¼ ë¶™ì—¬ ì†ì‹¤ ê°’ì„ ì–‘ìˆ˜ë¡œ ë³€í™˜í•˜ì—¬ ìµœì í™”ê°€ ì›í™œí•˜ê²Œ ì´ë£¨ì–´ì§€ë„ë¡ í•œë‹¤**  

ğŸš€ **ì¦‰, ê²½ì‚¬ í•˜ê°•ë²•ì€ ì†ì‹¤ í•¨ìˆ˜ì˜ ê¸°ìš¸ê¸°ë¥¼ ì´ìš©í•´ ì ì§„ì ìœ¼ë¡œ ìµœì ì ì„ ì°¾ëŠ” í•µì‹¬ í•™ìŠµ ë°©ë²•ì´ë‹¤!**

````python
#ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì™€ì„œ í™•ë¥ ì  ê²½ì‚¬ í•˜ê°•ë²•ì„ ì ìš©í•´ë³´ì.

import pandas as pd

df = pd.read_csv('data/Fish.csv')
````

````python
fish_input = df[['Weight', 'Length2', 'Length3', 'Height', 'Width']]
fish_target = df[['Species']]
````

````python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = \
train_test_split(fish_input, fish_target)
````

````python
from sklearn.preprocessing import StandardScaler
# ì •ê·œí™”ê¹Œì§€
ss = StandardScaler()
ss.fit(train_input)

train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
````

````python
# SGDClassifier 
# í™•ë¥ ì  ê²½ì‚¬í•˜ê°•ë²• ì ìš©í•˜ê¸°
from sklearn.linear_model import SGDClassifier

sc = SGDClassifier(loss='log_loss', max_iter=100)
# ì±…ì— ë‚˜ì˜¨ ë‚´ìš©ê³¼ ë™ì¼í•˜ê²Œ ì½”ë”©
# log_lossë¥¼ í†µí•´ ìŒìˆ˜í™”í•œë‹¤ëŠ” ì–˜ê¸°, max_iterëŠ” 10ë²ˆì˜ epoke(ì—í¬í¬)ë¥¼ ì‹¤í–‰í•œë‹¤ => 10ë²ˆì˜ ê·¸ë˜í”„ë¥¼ ë§Œë“ ë‹¤. 
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))

# ì§€ê¸ˆí•˜ê³  ìˆëŠ” ë°ì´í„°ë¥¼ ê¸°ë°˜ìœ¼ë¡œ í•œ ë²ˆ ë” í•™ìŠµí•´
# => ë¼ëŠ” ì½”ë“œë¥¼ ì•„ë˜ì— ì´ì–´ì„œ ì¨ë³´ê² ìŒ.
````

````python
sc.partial_fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))

# epokeë¥¼ ë”í•œë‹¤ê³  í•´ì„œ (ê·¸ë˜í”„ë¥¼ ë” ë§ì´ ê·¸ë ¤ë³¸ë‹¤ê³  í•´ì„œ) ì •í™•ë„ê°€ ì—„ì²­ë‚˜ê²Œ ë†’ì•„ì§€ì§€ëŠ” ì•ŠëŠ”ë‹¤. (p.211)
````
