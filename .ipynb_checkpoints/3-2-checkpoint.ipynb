# 03-2 선형 회귀 (Linear regression)

> (p.130)
> 데이터를 가장 잘 대변하는 최적의 선을 찾은 과정 

```python
import pandas as pd

df = pd.read_csv('data/Fish.csv')
df.head()
```

```python
# Perch 데이터 사용
perch_df = df.loc[df['Species'] == 'Perch']

#길이를 기반으로 무게를 예측하기
perch_length = perch_df[['Length2']]
perch_weight = perch_df[['Weight']]

  # 모델 학습을 위해 데이터 프레임을 2차원으로 => 대괄호를 두번 붙인다.
print(perch_length, perch_weight)
```

```python
from sklearn.model_selection import train_test_split

#test,train 데이터 (4종류) 로 나누기 

train_input, test_input, train_target, test_target = \
train_test_split(perch_length, perch_weight)
```

```python
# 선형회귀 사용하기
# 선형 회귀 모델 훈련 
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(train_input, train_target)
```

```python
lr.score(test_input, test_target)
```

```python
lr.coef_
# coefficient 는 가중치 (weight)라고 부름.
# 영향을 얼만큼 줄까하는 정도 (머신러닝에서의 기울기)
```

```python
lr.intercept_
```

```python
# 시각화하기
import matplotlib.pyplot as plt

plt.scatter(train_input, train_target)
plt.plot([15, 50], [15 * 39 - 726, 50 * 39 - 726])
#plot은 선그래프를 그려줌 (시작점과 끝점)
#최적의 회귀선을 그림 (직선 plot으로)
```

```python
print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))
```

## 다항회귀 (p.139)

```python
lr.predict([[10]])
# 10mm일 때 => 음수의 무게를 가진다? 
# 말이 안되는 가정
# 선형회귀의 문제점 : 1차원의 그래프로는 표현을 할 수 없는 데이터들이 발생한다.

# 따라서 '다항회귀' 를 통해 곡선 형태의 선을 그려, 데이터에 적합한 형태의 회귀선을 그리자.
# 2차원 그래프를 그리자.
```

```python
train_input.head()
#학습 데이터에 뭐가 있는지 먼저 확인해보자
```

```python
train_input['Length2_poly'] = train_input['Length2'] ** 2
# 제곱을 시키자, 그리고 그 데이터를 다시 train_input에 새로운 컬럼 'Length2_poly'로 추가하자.

test_input['Length2_poly'] = test_input['Length2'] ** 2
train_input.head()
```

```python
# 다시 학습시켜보자 (다항회귀 - 제곱값으로)
lr = LinearRegression()
lr.fit(train_input, train_target)
```

```python
lr.score(test_input, test_target)
```

```python
lr.predict([[2, 4]])
```

```python
print(lr.coef_, lr.intercept_)
# 이번에는 2차방정식이므로, [리스트 안에 2개의 값], 값 이 출력됨.
```
