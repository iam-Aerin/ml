```python
import pandas as pd

df = pd.read_csv('http://bit.ly/wine_csv_data')
data = df[['alcohol', 'sugar', 'pH']]
target = df['class']
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target =\
train_test_split(data, target)
```

## `RandomForestClassifier`
#### ëžœë¤í¬ë ˆìŠ¤íŠ¸ëŠ” ê²°ì •íŠ¸ë¦¬ì˜ ì•™ìƒë¸”ì´ê¸° ë–„ë¬¸ì— 
- DecisionTreeClassifier ê°€ ì œê³µí•˜ëŠ” ì¤‘ìš”í•œ ë§¤ê°œ ë³€ìˆ˜ë¥¼ ëª¨ë‘ ì œê³µí•¨.
- 
https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html

```python
# ëžœë¤í¬ë ˆìŠ¤íŠ¸ ì‹¤í–‰
# RandomForestClassifier

from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_jobs=1)
scores = cross_validate(rf, train_input, train_target, return_train_score=True, n_jobs=1)
print(scores)

# cross_validateëŠ” êµì°¨ ê²€ì¦ì„ ìˆ˜í–‰í•˜ì—¬ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ í‰ê°€í•˜ëŠ” í•¨ìˆ˜ì´ë‹¤.
# return_train_score=True ì˜µì…˜ì„ ì‚¬ìš©í•˜ë©´ í›ˆë ¨ ì„¸íŠ¸ ì ìˆ˜ë„ í•¨ê»˜ ì œê³µëœë‹¤.
# ê²°ê³¼ëŠ” fit_time, score_time, test_score, train_score ë“±ì„ í¬í•¨í•˜ëŠ” ë”•ì…”ë„ˆë¦¬ í˜•íƒœë¡œ ë°˜í™˜ëœë‹¤.
#
# ðŸ’¡ ì´ ì½”ë“œì˜ ëª©ì ì€ ëžœë¤ í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ì´ í›ˆë ¨ ë°ì´í„°ì™€ ê²€ì¦ ë°ì´í„°ì—ì„œ ì–¼ë§ˆë‚˜ ìž˜ ìž‘ë™í•˜ëŠ”ì§€ë¥¼ í™•ì¸í•˜ëŠ” ê²ƒ!
```

```python
# ë°©ê¸ˆ í•™ìŠµí•œ ëžœë¤í¬ë ˆìŠ¤íŠ¸ì—; feature_importances_ í™•ì¸í•˜ê¸°

rf.fit(train_input, train_target)
rf.feature_importances_

# í›ˆë ¨ ì„¸íŠ¸ì— í›ˆë ¨í•œ í›„ 'íŠ¹ì„± ì¤‘ìš”ë„'ë¥¼ ì¶œë ¥í•¨. |

# ê°ê° [ì•Œì½”ì˜¬, ë„ìˆ˜, ë‹¹ë„, pH] ì˜€ìœ¼ë¯€ë¡œ, ê²°ê³¼ê°’ì„ ì½ì—ˆì„ ë•Œ, ë‘ë²ˆì§¸ì¸ 'ë‹¹ë„'ì˜ ì¤‘ìš”ë„ê°€ í™”ì´íŠ¸ì™€ ë ˆë“œë¥¼ ë‚˜ëˆ„ëŠ”ë°ì— ê°€ìž¥ ì¤‘ìš”í•œ ê²ƒ ê°™ë‹¤~
# íŠ¹ì„±ì„ ëžœë¤í•˜ê²Œ ì„ íƒí–ˆê¸° ë•Œë¬¸ì— (ë‚´ê°€ ì“°ê³  ì‹¶ì€ íŠ¹ì„±ì„ ì„ íƒí•´ì„œ ë§Œë“¤ì–´ì„œ) => ê³¼ëŒ€ì í•©ì„ ì¤„ì´ê³  ì¼ë°˜í™”ë¥¼ í•  ìˆ˜ ìžˆë‹¤.
```

```python
# ì²œ ê°œ ì¤‘ì— ëžœë¤í•˜ê²Œ ë°± ê°œë¥¼ ë½‘ì•„ ì˜¬ ìˆ˜ë„ ìžˆê³ , ì‚¼ì‹­ ê°œë§Œ ë½‘ì•„ ì˜¬ ìˆ˜ë„ ìžˆë‹¤. 
# ê·¸ëŸ° ê³¼ì •ì—ì„œ í•œ ë²ˆë„ ì„ íƒë°›ì§€ ëª»í•œ ë°ì´í„°ê°€ ìžˆì„ ìˆ˜ ìžˆë‹¤.

# ê·¸ ì„ íƒ ë°›ì§€ ëª»í•œ ë°ì´í„°ë¥¼
# OOB ìƒ˜í”Œì´ë¼ê³  ë¶€ë¥¸ë‹¤.
# ì´ëŠ” í•œë²ˆë„ í›ˆë ¨ì— í¬í•¨ë˜ì§€ ì•Šì•˜ê¸° ë•Œë¬¸ì— ë‚˜ì¤‘ì— í…ŒìŠ¤íŠ¸ì— í™œìš©í•  ìˆ˜ ìžˆë‹¤. 

# OOB ì ìˆ˜ ì¶œë ¥í•˜ê¸°
rf= RandomForestClassifier(oob_score=True, n_jobs=1)
rf.fit(train_input, train_target)
rf.oob_score_
```

> ë…¸ë“œë¥¼ ë¶„í• í•  ë•Œ ê°€ìž¥ ì¢‹ì€ ë¶„í• ì„ ì°¾ëŠ” ê²ƒì´ ì•„ë‹ˆë¼ `ë¬´ìž‘ìœ„ë¡œ ë¶„í• `!

```python
from sklearn.ensemble import ExtraTreesClassifier
et = ExtraTreesClassifier(n_jobs=1)

scores= cross_validate(et, train_input, train_target, return_train_score=True, n_jobs=1)
print(scores)

#ëžœë¤í¬ë ˆìŠ¤íŠ¸ì™€ ë¹„ìŠ·í•œ ê²°ê³¼ê°’ì„ ë³´ì—¬ì¤Œ.
# ìš°ë¦¬ì˜ ë°ì´í„°ê°€ ë§Žì§€ê°€ ì•Šì•„ì„œ~
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_validate

# ëª¨ë¸ ê°ì²´ ìƒì„±
# gb = GradientBoostingClassifier(n_estimators=500)
# ì‹¬ì„ ë‚˜ë¬´ì˜ ê°œìˆ˜ë¥¼ ì§€ì •í•´ ì¤„ ìˆ˜ë„ ìžˆìŒ

gb = GradientBoostingClassifier()


# êµì°¨ ê²€ì¦ ì‹¤í–‰
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=1)

# ê²°ê³¼ ì¶œë ¥
print(scores)
```

```python
from sklearn.inspection import permutation_importance

hgb.fit(train_input, train_target)
scores = permutation_importance(hgb, train_input, train_target)
print(scores.importances_mean)

# ë‹¹ë„ ì ìˆ˜ë¥¼ ì„žì—ˆì„ ë•Œ, ìˆ˜ì¹˜ê°€ ë§Žì´ ë‚®ì•„ì§€ë„¤??
```
