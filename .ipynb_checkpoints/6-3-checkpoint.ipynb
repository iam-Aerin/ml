#### 주성분 분석 (PCA, Principal Component Analysis)

`"PCA는 데이터의 분산을 보존하는 방향으로 차원을 축소하여, 원본 특성을 선형 결합한 새로운 주성분을 생성한다."`

> 📌 1. 주성분 분석(PCA)의 개념
데이터의 **분산(variance)이 가장 큰 방향**을 새로운 축으로 정하고, 이를 **주성분(Principal Components, PC)** 이라고 합니다.  
차원 축소를 통해 **데이터를 더 적은 변수로 표현**할 수 있으며,  
데이터의 핵심 정보를 유지하면서도 **노이즈를 제거**할 수 있습니다.  

PCA는 **비지도 학습 (Unsupervised Learning) 기법** 중 하나 

- 비슷한 특성을 가진 데이터들을 비슷한 개념으로 (경향성을 띄는 데이터들에 대해) 축소 할 수 있을 것이다. 


````python
import numpy as np
fruits = np.load('data/fruits_300.npy')
fruits_2d = fruits.reshape(-1, 100*100)
````

````python
# 주성분 분석을 해주는 클래스 
# = PCA 클래스 (p.322)
from sklearn.decomposition import PCA

pca = PCA(n_components=50)
# 10000개의 픽셀을 50개로 줄여봐~ 하고 학습시킴

pca.fit(fruits_2d)
````

````python
pca.components_.shape
#PCA클래스가 찾은 주성분은 components_속성에 저장되어 있음. 
# n_components=50으로 지정했기 떄문에 pca.components_ 배열의 첫 번째 차원은 50
# 즉, 50개의 주성분을 찾은 상태
````

````python
import matplotlib.pyplot as plt

def draw_fruits(arr, ratio=1):
    n = len(arr)

    rows = int(np.ceil(n/10))
    cols = n if rows < 2 else 10

    fig, axs = plt.subplots(rows, cols, figsize=(cols*ratio, rows*ratio), squeeze=False)

    for i in range(rows):
        for j in range(cols):
            if i*10 + j < n:
                axs[i, j].imshow(arr[i*10 + j], cmap='gray_r')
            axs[i, j].axis('off')
    plt.show()

# 우리가 가진 데이터를 그림으로 만들어주는 코드 (복사 붙여넣기함) => 다음 코드에 필요해서
````

````python
draw_fruits(pca.components_.reshape(-1, 100, 100))
# 다시 그림으로 바꿔주려고 .reshape() 함
````

````python
# 지금까지 학습을 시켰고, transform을 해보겠다. 
# 원본 데이터를 주성분 분석에 투영 시키겠다. 

fruits_pca = pca.transform(fruits_2d)        
fruits_pca.shape
````

가장 중요하다고 생각하는 픽셀 50개로 압축을 시킴. 
데이터를 성공적으로 줄임 ! 
=> 1/200 만큼 (원래가 10,000개 였으니)

---

압축한 것을 이제 다시 원본 데이터로 바꿔보자!

````python
fruits_inverse = pca.inverse_transform(fruits_pca)
f = fruits_inverse.reshape(-1, 100, 100)
draw_fruits(f[:100])

# 압축했다가 다시 원본화하고 (원상복구), 그걸 그림으로 그려줘

# 50개의 특성을 다시 10,000개로 늘림에도 과일이 잘 복원됨. 
````

````python
np.sum(pca.explained_variance_ratio_)
# 94%가 넘는 분산을 유지하고 있음. 
````

````python
pca.explained_variance_ratio_
````

````python
# 로지스틱 회귀를 써보자
# 지도학습
# 타겟을 알려줘야함.


from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
# 원본데이터를 (10000개 짜리) 학습시킴.
target = np.array([0]*100 + [1]*100 + [2]*100)
# 지금은 알기 때문에 이렇게 타겟을 지정해서 만듦.
````

````python
from sklearn.model_selection import cross_validate
scores = cross_validate(lr, fruits_2d, target)
print(scores)

# 걸린 시간: 0.4836초 정도 걸림
# 데이터는 많이 맞춘 것으로 보임.
````

````python
scores = cross_validate(lr, fruits_pca, target)
print(scores)

# 이번에는 원본 데이터가 아니라 frutis_pca => 50개로 차원 축소한 데이터를 넣음. 
# 시간이 단축 된 것을 확인 할 수 있음. 
# 걸린 시간: 0.028초
````

````python
# 주성분 분석을 할 때, 조금 더 적은 데이터를 써! 
# 다시 한 번 학습해봐 

pca = PCA(n_components=0.5)
# 설명된 분산에 50%에 달하는 주성분을 찾도록
pca.fit(fruits_2d)
pca.n_components_
````

````python
fruits_pca = pca.transform(fruits_2d)
fruits_pca.shape
# 주성분이 2개로 줄은 데이터
````

````python
scores = cross_validate(lr, fruits_pca, target)
scores

# 지금은 완벽한 데이터라서 수치가 굉장히 높게 나타남
# 주성분이 2개인 상태에서도
````
