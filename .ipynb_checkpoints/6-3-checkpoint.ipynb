#### Ï£ºÏÑ±Î∂Ñ Î∂ÑÏÑù (PCA, Principal Component Analysis)

`"PCAÎäî Îç∞Ïù¥ÌÑ∞Ïùò Î∂ÑÏÇ∞ÏùÑ Î≥¥Ï°¥ÌïòÎäî Î∞©Ìñ•ÏúºÎ°ú Ï∞®ÏõêÏùÑ Ï∂ïÏÜåÌïòÏó¨, ÏõêÎ≥∏ ÌäπÏÑ±ÏùÑ ÏÑ†Ìòï Í≤∞Ìï©Ìïú ÏÉàÎ°úÏö¥ Ï£ºÏÑ±Î∂ÑÏùÑ ÏÉùÏÑ±ÌïúÎã§."`

> üìå 1. Ï£ºÏÑ±Î∂Ñ Î∂ÑÏÑù(PCA)Ïùò Í∞úÎÖê
Îç∞Ïù¥ÌÑ∞Ïùò **Î∂ÑÏÇ∞(variance)Ïù¥ Í∞ÄÏû• ÌÅ∞ Î∞©Ìñ•**ÏùÑ ÏÉàÎ°úÏö¥ Ï∂ïÏúºÎ°ú Ï†ïÌïòÍ≥†, Ïù¥Î•º **Ï£ºÏÑ±Î∂Ñ(Principal Components, PC)** Ïù¥ÎùºÍ≥† Ìï©ÎãàÎã§.  
Ï∞®Ïõê Ï∂ïÏÜåÎ•º ÌÜµÌï¥ **Îç∞Ïù¥ÌÑ∞Î•º Îçî Ï†ÅÏùÄ Î≥ÄÏàòÎ°ú ÌëúÌòÑ**Ìï† Ïàò ÏûàÏúºÎ©∞,  
Îç∞Ïù¥ÌÑ∞Ïùò ÌïµÏã¨ Ï†ïÎ≥¥Î•º Ïú†ÏßÄÌïòÎ©¥ÏÑúÎèÑ **ÎÖ∏Ïù¥Ï¶àÎ•º Ï†úÍ±∞**Ìï† Ïàò ÏûàÏäµÎãàÎã§.  

PCAÎäî **ÎπÑÏßÄÎèÑ ÌïôÏäµ (Unsupervised Learning) Í∏∞Î≤ï** Ï§ë ÌïòÎÇò 

- ÎπÑÏä∑Ìïú ÌäπÏÑ±ÏùÑ Í∞ÄÏßÑ Îç∞Ïù¥ÌÑ∞Îì§ÏùÑ ÎπÑÏä∑Ìïú Í∞úÎÖêÏúºÎ°ú (Í≤ΩÌñ•ÏÑ±ÏùÑ ÎùÑÎäî Îç∞Ïù¥ÌÑ∞Îì§Ïóê ÎåÄÌï¥) Ï∂ïÏÜå Ìï† Ïàò ÏûàÏùÑ Í≤ÉÏù¥Îã§. 


````python
import numpy as np
fruits = np.load('data/fruits_300.npy')
fruits_2d = fruits.reshape(-1, 100*100)
````

````python
# Ï£ºÏÑ±Î∂Ñ Î∂ÑÏÑùÏùÑ Ìï¥Ï£ºÎäî ÌÅ¥ÎûòÏä§ 
# = PCA ÌÅ¥ÎûòÏä§ (p.322)
from sklearn.decomposition import PCA

pca = PCA(n_components=50)
# 10000Í∞úÏùò ÌîΩÏÖÄÏùÑ 50Í∞úÎ°ú Ï§ÑÏó¨Î¥ê~ ÌïòÍ≥† ÌïôÏäµÏãúÌÇ¥

pca.fit(fruits_2d)
````

````python
pca.components_.shape
#PCAÌÅ¥ÎûòÏä§Í∞Ä Ï∞æÏùÄ Ï£ºÏÑ±Î∂ÑÏùÄ components_ÏÜçÏÑ±Ïóê Ï†ÄÏû•ÎêòÏñ¥ ÏûàÏùå. 
# n_components=50ÏúºÎ°ú ÏßÄÏ†ïÌñàÍ∏∞ ÎñÑÎ¨∏Ïóê pca.components_ Î∞∞Ïó¥Ïùò Ï≤´ Î≤àÏß∏ Ï∞®ÏõêÏùÄ 50
# Ï¶â, 50Í∞úÏùò Ï£ºÏÑ±Î∂ÑÏùÑ Ï∞æÏùÄ ÏÉÅÌÉú
````

````python
import matplotlib.pyplot as plt

def draw_fruits(arr, ratio=1):
    n = len(arr)

    rows = int(np.ceil(n/10))
    cols = n if rows < 2 else 10

    fig, axs = plt.subplots(rows, cols, figsize=(cols*ratio, rows*ratio), squeeze=False)

    for i in range(rows):
        for j in range(cols):
            if i*10 + j < n:
                axs[i, j].imshow(arr[i*10 + j], cmap='gray_r')
            axs[i, j].axis('off')
    plt.show()

# Ïö∞Î¶¨Í∞Ä Í∞ÄÏßÑ Îç∞Ïù¥ÌÑ∞Î•º Í∑∏Î¶ºÏúºÎ°ú ÎßåÎì§Ïñ¥Ï£ºÎäî ÏΩîÎìú (Î≥µÏÇ¨ Î∂ôÏó¨ÎÑ£Í∏∞Ìï®) => Îã§Ïùå ÏΩîÎìúÏóê ÌïÑÏöîÌï¥ÏÑú
````

````python
draw_fruits(pca.components_.reshape(-1, 100, 100))
# Îã§Ïãú Í∑∏Î¶ºÏúºÎ°ú Î∞îÍøîÏ£ºÎ†§Í≥† .reshape() Ìï®
````

````python
# ÏßÄÍ∏àÍπåÏßÄ ÌïôÏäµÏùÑ ÏãúÏº∞Í≥†, transformÏùÑ Ìï¥Î≥¥Í≤†Îã§. 
# ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞Î•º Ï£ºÏÑ±Î∂Ñ Î∂ÑÏÑùÏóê Ìà¨ÏòÅ ÏãúÌÇ§Í≤†Îã§. 

fruits_pca = pca.transform(fruits_2d)        
fruits_pca.shape
````

Í∞ÄÏû• Ï§ëÏöîÌïòÎã§Í≥† ÏÉùÍ∞ÅÌïòÎäî ÌîΩÏÖÄ 50Í∞úÎ°ú ÏïïÏ∂ïÏùÑ ÏãúÌÇ¥. 
Îç∞Ïù¥ÌÑ∞Î•º ÏÑ±Í≥µÏ†ÅÏúºÎ°ú Ï§ÑÏûÑ ! 
=> 1/200 ÎßåÌÅº (ÏõêÎûòÍ∞Ä 10,000Í∞ú ÏòÄÏúºÎãà)

---

ÏïïÏ∂ïÌïú Í≤ÉÏùÑ Ïù¥Ï†ú Îã§Ïãú ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞Î°ú Î∞îÍøîÎ≥¥Ïûê!

````python
fruits_inverse = pca.inverse_transform(fruits_pca)
f = fruits_inverse.reshape(-1, 100, 100)
draw_fruits(f[:100])

# ÏïïÏ∂ïÌñàÎã§Í∞Ä Îã§Ïãú ÏõêÎ≥∏ÌôîÌïòÍ≥† (ÏõêÏÉÅÎ≥µÍµ¨), Í∑∏Í±∏ Í∑∏Î¶ºÏúºÎ°ú Í∑∏Î†§Ï§ò

# 50Í∞úÏùò ÌäπÏÑ±ÏùÑ Îã§Ïãú 10,000Í∞úÎ°ú ÎäòÎ¶ºÏóêÎèÑ Í≥ºÏùºÏù¥ Ïûò Î≥µÏõêÎê®. 
````

````python
np.sum(pca.explained_variance_ratio_)
# 94%Í∞Ä ÎÑòÎäî Î∂ÑÏÇ∞ÏùÑ Ïú†ÏßÄÌïòÍ≥† ÏûàÏùå. 
````

````python
pca.explained_variance_ratio_
````

````python
# Î°úÏßÄÏä§Ìã± ÌöåÍ∑ÄÎ•º Ïç®Î≥¥Ïûê
# ÏßÄÎèÑÌïôÏäµ
# ÌÉÄÍ≤üÏùÑ ÏïåÎ†§Ï§òÏïºÌï®.


from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
# ÏõêÎ≥∏Îç∞Ïù¥ÌÑ∞Î•º (10000Í∞ú ÏßúÎ¶¨) ÌïôÏäµÏãúÌÇ¥.
target = np.array([0]*100 + [1]*100 + [2]*100)
# ÏßÄÍ∏àÏùÄ ÏïåÍ∏∞ ÎïåÎ¨∏Ïóê Ïù¥Î†áÍ≤å ÌÉÄÍ≤üÏùÑ ÏßÄÏ†ïÌï¥ÏÑú ÎßåÎì¶.
````

````python
from sklearn.model_selection import cross_validate
scores = cross_validate(lr, fruits_2d, target)
print(scores)

# Í±∏Î¶∞ ÏãúÍ∞Ñ: 0.4836Ï¥à Ï†ïÎèÑ Í±∏Î¶º
# Îç∞Ïù¥ÌÑ∞Îäî ÎßéÏù¥ ÎßûÏ∂ò Í≤ÉÏúºÎ°ú Î≥¥ÏûÑ.
````

````python
scores = cross_validate(lr, fruits_pca, target)
print(scores)

# Ïù¥Î≤àÏóêÎäî ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞Í∞Ä ÏïÑÎãàÎùº frutis_pca => 50Í∞úÎ°ú Ï∞®Ïõê Ï∂ïÏÜåÌïú Îç∞Ïù¥ÌÑ∞Î•º ÎÑ£Ïùå. 
# ÏãúÍ∞ÑÏù¥ Îã®Ï∂ï Îêú Í≤ÉÏùÑ ÌôïÏù∏ Ìï† Ïàò ÏûàÏùå. 
# Í±∏Î¶∞ ÏãúÍ∞Ñ: 0.028Ï¥à
````

````python
# Ï£ºÏÑ±Î∂Ñ Î∂ÑÏÑùÏùÑ Ìï† Îïå, Ï°∞Í∏à Îçî Ï†ÅÏùÄ Îç∞Ïù¥ÌÑ∞Î•º Ïç®! 
# Îã§Ïãú Ìïú Î≤à ÌïôÏäµÌï¥Î¥ê 

pca = PCA(n_components=0.5)
# ÏÑ§Î™ÖÎêú Î∂ÑÏÇ∞Ïóê 50%Ïóê Îã¨ÌïòÎäî Ï£ºÏÑ±Î∂ÑÏùÑ Ï∞æÎèÑÎ°ù
pca.fit(fruits_2d)
pca.n_components_
````

````python
fruits_pca = pca.transform(fruits_2d)
fruits_pca.shape
# Ï£ºÏÑ±Î∂ÑÏù¥ 2Í∞úÎ°ú Ï§ÑÏùÄ Îç∞Ïù¥ÌÑ∞
````

````python
scores = cross_validate(lr, fruits_pca, target)
scores

# ÏßÄÍ∏àÏùÄ ÏôÑÎ≤ΩÌïú Îç∞Ïù¥ÌÑ∞ÎùºÏÑú ÏàòÏπòÍ∞Ä ÍµâÏû•Ìûà ÎÜíÍ≤å ÎÇòÌÉÄÎÇ®
# Ï£ºÏÑ±Î∂ÑÏù¥ 2Í∞úÏù∏ ÏÉÅÌÉúÏóêÏÑúÎèÑ
````
